```python
import numpy as np
from matplotlib import pyplot as plt
import plotly as pl
import plotly.graph_objs as go
```

```python
def Fn(x, X):
    count = 0
    for xx in X:
        if xx<=x:
            count+=1
    return count/len(X)
```

```python
def generate_X(X, mode = 0, n = 3):
    if mode == 0:
        X[:] = [0.0]
        for i in range(n):
            X.append(np.random.rand())
        X.append(1.0)
    if mode == 1:
        X.append(np.random.rand())
    X.sort()
```

```python
X = []
```

```python
generate_X(X, mode = 0)
Y = [i * 1/(len(X)-2) for i in range(len(X)-1)] + [1.0]
#Y = [(i+1) * 1/(len(X)) for i in range(len(X))]
```
